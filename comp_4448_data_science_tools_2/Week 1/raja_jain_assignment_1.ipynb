```python
# Read the boston dataset csv files provided for this assignment into Python (you can use pd.read_csv()).

# The boston datasets are boston_weather_1, boston_weather_2, boston_weather_3, boston_weather_4, and more_weather_variables, then assign the datasets to a DataFrame variable called boston_1, boston_2, boston_3, boston_4, and more_variables, respectively.

boston_1 = pd.read_csv("boston_weather_1.csv")
boston_2 = pd.read_csv("boston_weather_2.csv")
boston_3 = pd.read_csv("boston_weather_3.csv")
boston_4 = pd.read_csv("boston_weather_4.csv")
```

```python
# Combine or concatenate the DataFrames, boston_1, boston_2, boston_3, and boston_4 and assign the results to a DataFrame called combined_boston.

# These four datasets should be combined vertically since they have the same variable names, such that boston_1 is stacked on top of boston_2, and the result is stacked on top of boston_3, and the result is further stacked on top of boston_4.

combined_boston = pd.concat([boston_1, boston_2, boston_3, boston_4])
combined_boston.reset_index(drop=True, inplace=True)
combined_boston.head()
```

```python
# Horizontally merge, join or concatenate the combined_boston and more_variables DataFrames and assign the results to a DataFrame called boston_data.

more_variables = pd.read_csv("more_weather_variables.csv")

boston_data = pd.concat([combined_boston, more_variables], axis=1)
```

```python
# Print the first five rows of the boston_data, and the last five rows of the boston_data. Also print out the shape of the boston_data.
print("First five rows of the boston_data")
display(boston_data.head())

print("Last five rows of the boston_data")
display(boston_data.tail())

print("boston_data_shape:", boston_data.shape)
```

```python
# Check the boston_data to verify how many missing data points exist under each column.
display(boston_data.isnull().sum())
```

```python
# Drop the rows or instances that contain any missing data.
# Assign the resulting DataFrame to a variable called clean_boston_data.
clean_boston_data = boston_data.dropna()
```

```python
# Check for missing data again to ensure there is no missing data in the clean_boston_data.
display(clean_boston_data.isnull().sum())
```

```python
# Print the shape of the clean_boston_data
print("clean_boston_data_shape:", clean_boston_data.shape)
```

```python
# Format all the column names to lowercase and include underscore between column names that consist of two words.
print("clean_boston_data_columns:", clean_boston_data.columns)


def camel_case_split(str):
    """returns a list of individual words from camel case string"""
    start_idx = [i for i, e in enumerate(str) if e.isupper()] + [len(str)]

    start_idx = [0] + start_idx
    return [str[x:y] for x, y in zip(start_idx, start_idx[1:])]
```

```python
# Reassign the DataFrame with the formatted column names to the same variable, clean_boston_data. Print or output the columns of the clean_boston_data DataFrame.
clean_boston_data.columns = [
    "_".join(camel_case_split(col)).lower().strip("_")
    for col in clean_boston_data.columns
]

print("clean_boston_data_columns:", clean_boston_data.columns)
```

```python
# Select or slice all data from the clean_boston_data DataFrame, except the data where the Year is 1930. You can call this subset data excluding_1930.
excluding_1930 = clean_boston_data[clean_boston_data["year"] != 1930]
```

```python
# Using the excluding_1930 DataFrame, output the first 20 unique values in the Year column.
print(
    "first 20 unique values in the Year column:", excluding_1930["year"].unique()[:20]
)
```

```python
# Select the data from the clean_boston_data where the Year is and 1995 AND the high_temp is greater than or equal to 90. Output or display the whole selected data. Here, you don’t have to assign it to any variable, but you could if you want to.
print("data where the Year is 1995 AND the high_temp is greater than or equal to 90:")
display(
    clean_boston_data[
        (clean_boston_data["year"] == 1995) & (clean_boston_data["high_temp"] >= 90)
    ]
)
```

```python
# Select the data from the clean_boston_data where the Year is and 1995 OR the high_temp is greater than 89. Output or print the first 20 rows of the selected data. Here, you don’t have to assign it to any variable, but you could if you want to.
print("data where the Year is 1995 OR the high_temp is greater than 89:")
display(
    clean_boston_data[
        (clean_boston_data["year"] == 1995) | (clean_boston_data["high_temp"] > 89)
    ][:20]
)
```

```python
# Read the student_data file provided into Python (take note of the file extension to use the appropriate pandas reader to read the data).
student_data = pd.read_excel("student_data.xlsx")
display(student_data.head())
```

```python
# Drop the first empty column in Python and assign the DataFrame to a variable student_data.
student_data = student_data.drop(student_data.columns[0], axis=1)
display(student_data.head())
```

```python
# Clean student_data column names
print(student_data.columns)
student_data.columns = [col.strip() for col in student_data.columns]
print(student_data.columns)
```

```python
# Clean student_data instance data
display(student_data.info())
```

```python
# strip spaces from student_data['ice_cream_flavor']
student_data["ice_cream_flavor"] = student_data["ice_cream_flavor"].str.strip()
```

```python
# replace chocollate in student_data['ice_cream_flavor'] with chocolate
student_data["ice_cream_flavor"] = student_data["ice_cream_flavor"].str.replace(
    "chocollate", "chocolate"
)
```

```python
# The student_data shows the different midterm scores of students in math, reading and science, and their favorite ice cream flavors.
# Select the data in the ice_cream_flavor column and convert the flavors to a numpy array, then assign it to a variable called flavor.
flavor = student_data["ice_cream_flavor"].values
```

```python
# From the student_data, select the math, reading and science scores all at once and convert the selected data to a numpy array and assign it to a variable called scores.
scores = student_data[["math", "reading", "science"]].values
```

```python
# Print the data in the flavor and scores arrays.
print("flavor")
display(flavor)
print("scores")
display(scores)
```

```python
# Use the scores and flavor arrays to slice out the scores where the flavor is chocolate only.
print("chocolate scores:\n", scores[np.where(flavor == "chocolate")])
```

```python
# Use the scores and flavor arrays to slice out the scores where the flavor is chocolate OR vanilla.
print(
    "chocolate or vanilla scores:\n",
    scores[np.where((flavor == "chocolate") | (flavor == "vanilla"))],
)
```

```python
# Use the scores and flavor arrays to slice out the scores where the flavor is not chocolate (you can use the “~ “sign).
print("not chocolate scores:\n", scores[np.where(~(flavor == "chocolate"))])
```

```python
# Slice out all math and reading scores where the flavor is chocolate, then compute the mean of math and reading scores for this subset (you can use the .mean() method on the array and specify the axis parameter value appropriately).
print(
    "mean of math and reading scores for chocolate:\n",
    scores[
        np.where((flavor == "chocolate") & (scores[:, 0] > 0) & (scores[:, 1] > 0))
    ].mean(axis=0),
)
```

```python
# Imagine that you wanted to use the student_data in question 2a to make predictions such that the ice_cream_flavor, math and reading columns are input variables and science column is the output variable you want to predict.

# Use the LabelBinarizer() in the sklearn package to transform the ice_cream_flavor column in the student_data to dummy variables, then join these dummy variables to the student_data and drop the original ice_cream_flavor column. Reassign the resulting DataFrame to a variable called student_data_1. Print out the entire student_data_1 DataFrame

display(student_data.head())
student_data_1 = pd.concat(
    [
        student_data,
        pd.DataFrame(LabelBinarizer().fit_transform(student_data["ice_cream_flavor"])),
    ],
    axis=1,
)
student_data_1.drop("ice_cream_flavor", axis=1, inplace=True)
student_data_1.columns = [
    "math",
    "reading",
    "science",
    "chocolate",
    "strawberry",
    "vanilla",
]
display(student_data_1)
```

```python
# Extract the math, reading and science scores and use the StandardScaler() class in the sklearn.preprocessing module to standardize these scores, then merge the standardized scores to the dummy variables, and call the resulting DataFrame student_data_std. Print out the entire student_data_std DataFrame.
student_data_std = student_data_1.copy()
student_data_std[["math", "reading", "science"]] = StandardScaler().fit_transform(
    student_data_1[["math", "reading", "science"]]
)
display(student_data_std.head())
```

```python
# Using a split ratio of 70:30, split the student_data_1 into training and test set.
# Reference the input and output of the training set as X_train and y_train respectively. Also reference the input and output of the test set as X_test and y_test respectively.
X_train, y_train, X_test, y_test = train_test_split(
    student_data_1.drop("science", axis=1),
    student_data_1["science"],
    test_size=0.3,
    random_state=1,
)
```

```python
# Print X_train, y_train, X_test and y_test.
display("X_train:", X_train)
display("X_test:", X_test)

display("y_train:", y_train)
display("y_test:", y_test)
```

```python
# Using the boston_1 DataFrame in question 1a, select the LowTemp, HighTemp, WarmestMin, ColdestHigh, AveMin, AveMax columns and use the “pipeline” functionality in sklearn to transform the selected data using the MinMaxScalar() and Imputer() classes.
# With the Imputer() class, the missing data in each column should be imputed using the mean value for the column.
num_proc = make_pipeline(SimpleImputer(strategy="mean"), MinMaxScaler())
```

```python
# Assign the resulting DataFrame to a variable called pipeline_data.
cols = ["LowTemp", "HighTemp", "WarmestMin", "ColdestHigh", "AveMin", "AveMax"]
pipeline_data = num_proc.fit_transform(boston_1[cols])
```

```python
# Print the pipeline_data
display("pipeline_data", pipeline_data)
```

```python
# Output the descriptive statistics of the pipeline_data including mean, median, variance, minimum value, maximum value, variance, standard deviation and skewness.
# Use the .apply() functionality of the pandas DataFrame.
display(
    "Descriptive Statistics",
    pd.DataFrame(pipeline_data, columns=cols).apply(
        ["mean", "median", "var", "min", "max", "std", "skew"], axis=0
    ),
)
```
