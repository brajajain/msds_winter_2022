g)	Create a function that computes the overall accuracy of the classification. You can call this function overall_accuracy. The function should take two arguments, a vector of  actual output values and a vector of predicted output values, then returns the classification accuracy. Inside the function, you can use boolean logic to compare the actual and predicted values, then count the proportion of how many predicted values are equal to the actual values to get the overall accuracy. 

```python
# Create a function that computes the overall accuracy of the classification. You can call this function overall_accuracy. The function should take two arguments, a vector of  actual output values and a vector of predicted output values, then returns the classification accuracy. Inside the function, you can use boolean logic to compare the actual and predicted values, then count the proportion of how many predicted values are equal to the actual values to get the overall accuracy.


def create_confusion_matrix(actual: np.ndarray, predicted: np.ndarray):
    """
    Computes the overall accuracy of the classification.
    """
    tn = 0
    fp = 0
    fn = 0
    tp = 0
    total = 0

    for a, p in zip(actual, predicted):
        total += 1
        if a == 0 and p == 0:
            tn += 1
        elif a == 0 and p == 1:
            fp += 1
        elif a == 1 and p == 0:
            fn += 1
        elif a == 1 and p == 1:
            tp += 1

    return [[tn, fp], [fn, tp]], total


def overall_accuracy(confusion_matrix: np.ndarray):
    tn, fp, fn, tp = (
        confusion_matrix[0][0],
        confusion_matrix[0][1],
        confusion_matrix[1][0],
        confusion_matrix[1][1],
    )
    return (tn + tp) / (tn + fp + fn + tp)


confusion_matrix, _ = create_confusion_matrix(
    actual=tax_data["cheat"].replace({"no": 0, "yes": 1}).values,
    predicted=tax_data["predicted_cheat"].values,
)

overall_accuracy(confusion_matrix)
```
